# 05. Define the Price label function

```python
def price_label(row):

  if row['prices'] <= 5:
     return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
       return 'Mid-range product'
  elif row['prices'] > 15:
       return 'High range'
  else: return np.nan
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

# Creating the "busiest_day" column

# Findout the frequency of orders_day_of_week

This information will tell which day the most orders took place

```python
# Print the frequency of the 'orders_day_of_week' column

orders_day_freq = ords_prods_merge['orders_day_of_week'].value_counts()
```

```python
print(orders_day_freq)
```

# From the printed frequencies, we know:

The busiest day (most orders) is Saturday, which corresponds to a value of 0.
    
The slowest day (least orders) is Wednesday, which corresponds to a value of 4.

#  Create the "busiest_day" Column Using a For-Loop

```python
# Create an empty list to store the labels

busiest_day_labels = []

# Loop through the 'orders_day_of_week' column

for day in ords_prods_merge['orders_day_of_week']:
    if day == 0:
        busiest_day_labels.append('Busiest day')
    elif day == 4:
        busiest_day_labels.append('Least busy')
    else:
        busiest_day_labels.append('Regularly busy')
```

```python
# Print the result

busiest_day_labels
```

# Create a new column within ords_prods_merge dataframe to combine the result

```python
# Add the new column to the dataframe

ords_prods_merge['busiest_day'] = busiest_day_labels
```

```python
ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

```python
# Display the first few rows to verify the new column

ords_prods_merge.head()
```

# Task 02. Update the "busiest_day" column

# Step 1: Identify the busiest and slowest days

```python
# Calculate the frequency of orders for each day of the week: (previously achieved)

# Identify the two busiest and two slowest days:

busiest_days = orders_day_freq.index[:2].tolist()  # First two days (busiest)

slowest_days = orders_day_freq.index[-2:].tolist()  # Last two days (slowest)
```

```python
# Print the busiest and slowest days

print("Two busiest days of the week:", busiest_days)

print("Two slowest days of the week:", slowest_days)
```

```python
# From the outcome;

# The two busiest days of the week are Saturday and Sunday, corresponding to 0 and 1

# The two slowest days of the week are Tuesday and Wednesday, corresponding to 3 and 4
```

# Step 2: Create the new column with updated labels

```python
# Create the new column based on these values

ords_prods_merge['busiest_days'] = ords_prods_merge['orders_day_of_week'].apply(
    lambda x: 'Busiest days' if x in busiest_days else ('Slowest days' if x in slowest_days else 'Regular days')
)
```

```python
# Display the first few rows to verify the new column

ords_prods_merge[['orders_day_of_week', 'busiest_days']].head()
```

# Task 03. Check the Values of the New Column to verify the accuracy of the new "busiest_days" column

```python
# Print the frequency of the new column to check for accuracy

busiest_days_freq = ords_prods_merge['busiest_days'].value_counts()

print("Frequency of the new 'busiest_days' column:\n", busiest_days_freq)
```

# Observation

The counts match the expected distribution of busiest, slowest, and regular days based on the initial frequency calculations

# Task 04: Identify the Busiest Periods of the Day

# Step 1. Calculate the frequency of orders for each hour:

* Sort these frequencies to categorize them.

```python
# Calculate the frequency of orders for each hour of the day

orders_hour_freq = ords_prods_merge['order_hour_of_day'].value_counts().sort_values(ascending=False)
```

```python
# Print the "Order frequencies by hour of the day:\n" 

orders_hour_freq
```

# Step 2. Determine thresholds for categorization

```python
# Determine thresholds for "Most orders", "Average orders", and "Fewest orders"

most_orders_hours = orders_hour_freq.index[:8].tolist()  # Top 8 hours

fewest_orders_hours = orders_hour_freq.index[-8:].tolist()  # Bottom 8 hours

# The remaining hours will be considered "Average orders"

average_orders_hours = orders_hour_freq.index[8:-8].tolist()
```

# Step 3. Create the new column, "busiest_period_of_day" Column

```python
ords_prods_merge['busiest_period_of_day'] = ords_prods_merge['order_hour_of_day'].apply(
    lambda x: 'Most orders' if x in most_orders_hours else ('Fewest orders' if x in fewest_orders_hours else 'Average orders')
)
```

```python
# Display the first few rows to verify the new column

ords_prods_merge[['order_hour_of_day', 'busiest_period_of_day']].head()
```
